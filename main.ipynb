### Social Network Ads Prediction using Logistic regression 

```python
# imports 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data

df = pd.read_csv('network_ad.csv')
df.head()
```

```python
# features

X = df[['Age', 'EstimatedSalary']].values
y = df['Purchased'].values
print(X.shape)
print(y.shape)
```

```python
# split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
# train the model

lr.fit(X_train, y_train)
```

```python
# prediction

y_pred = lr.predict(X_test)
```

##### blue = purchased (Clicked the ad)
##### red = not purchased (Did not click)

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='bwr')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Logistic Regression Prediction')
plt.show()
```

##### accuracy score of 90%

```python
from sklearn.metrics import accuracy_score

print("accuracy score: ", accuracy_score(y_pred, y_test)) # 90%
```
